```python
drop_features = ['name', 'host_name', 'host_location', 'host_about', 'host_has_profile_pic']
df.drop(columns=drop_features, inplace=True)
```

```python
#remove object types that can't get encoded
numeric_df = df.drop(columns=['description', 'neighborhood_overview', 'review_scores_rating' ], inplace=False)
numeric_df.dtypes
```

```python
missing = numeric_df.isnull().sum()
print(missing[missing > 0])
```

```python
#check if mean or median is better
print(numeric_df['host_response_rate'].mean())
print(numeric_df['host_response_rate'].median())

print(numeric_df['host_acceptance_rate'].mean())
print(numeric_df['host_acceptance_rate'].median())

print(numeric_df['bedrooms'].mean())
print(numeric_df['bedrooms'].median())

print(numeric_df['beds'].mean())
print(numeric_df['beds'].median())
```

```python
# Replace all NAN with the median
to_impute = ['host_response_rate', 'host_acceptance_rate', 'bedrooms', 'beds']
for feature in to_impute:
    numeric_df[feature].fillna(df[feature].median(), inplace=True)
missing = numeric_df.isnull().sum()
print(missing[missing > 0])
```

```python
# Winsorize the outliers
numeric_features = numeric_df.select_dtypes(include=[np.number])
winsorize_list = [feature for feature in numeric_features.columns if numeric_features[feature].nunique() > 5]
for feature in winsorize_list:
    numeric_df[feature] = stats.mstats.winsorize(numeric_df[feature], limits=(0.05, 0.95))
```

```python
# Standarize the features
scaler_standard = StandardScaler()
df_standardized = pd.DataFrame(scaler_standard.fit_transform(numeric_features), columns=numeric_features.columns)

scaler_normal = MinMaxScaler()
df_normalized = pd.DataFrame(scaler_normal.fit_transform(df_standardized), columns=df_standardized.columns)
```

```python
categorical_features = numeric_df.select_dtypes(include=[object])
categorical_features.drop(columns=['amenities'], inplace=True)
categorical_features.columns
```

```python
# Encode objects with low number of unique values
df_encoded = pd.get_dummies(categorical_features, columns=['neighbourhood_group_cleansed', 'room_type'])
df_encoded
```

```python
# Count up the amenities 
df_encoded['num_amenities'] = df['amenities'].apply(len)

average_amenities = df_encoded['num_amenities'].mean()
print("average:", average_amenities)
df_encoded
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
phrases = [item for item in unique_amenities if len(item.split()) > 1]


# Preprocess the amenities (ensure that amentities that are multi-worded stay together)
def preprocess_amenities(amenities, phrases):
    amenities_str = ' '.join(amenities)
    for phrase in phrases:
        amenities_str = amenities_str.replace(phrase, f"{phrase.replace(' ', '_')}")
    return amenities_str

# TF-IDF
vectorizer = TfidfVectorizer()
tfidf_matrix = vectorizer.fit_transform(df['amenities'].apply(lambda x: preprocess_amenities(x, phrases)))
feature_names = vectorizer.get_feature_names_out()
sum_tfidf = np.asarray(tfidf_matrix.sum(axis=0)).flatten()
tfidf_scores = dict(zip(feature_names, sum_tfidf))


top_10_amenities = sorted(tfidf_scores.items(), key=lambda x: x[1], reverse=True)[:10]
important_amenities = []

for amenity, score in top_10_amenities:
    print(f"{amenity}: {score:.4f}")
    revert_amenity = amenity.replace("_", " ")
    important_amenities.append(revert_amenity)
```

```python
encoded_amenities_df = pd.DataFrame()
for amenity in important_amenities:
    encoded_amenities_df[amenity] = 0

# One-hot encode the top ten amenities
for index, row in df.iterrows():
    amenities_list = [a.lower() for a in row['amenities']]
    for amenity in important_amenities:
        if amenity in amenities_list:
            encoded_amenities_df.at[index, amenity] = 1
        else:
            encoded_amenities_df.at[index, amenity] = 0

encoded_amenities_df
```

```python
combined_df = pd.concat([df_normalized, df_encoded, encoded_amenities_df], axis=1)
combined_df
```

```python
combined_df['instant_bookable'] = df['instant_bookable'].astype(int)
combined_df['has_availability'] = df['has_availability'].astype(int)
```

```python
combined_df['review_scores_rating'] = df['review_scores_rating']
corr_matrix = combined_df.corr()
target_corr = corr_matrix['review_scores_rating'].sort_values(ascending=False)
print(target_corr)
```

```python
threshold = 0.58
irrelevant_features = set()
for i in range(len(corr_matrix.columns)):
    if abs(corr_matrix.iloc[i, 2]) > threshold:
        colname = corr_matrix.columns[i]
        irrelevant_features.add(colname)

df_reduced = combined_df.drop(columns=irrelevant_features)

print("Original features:", combined_df.columns.tolist())
print("Features to drop:", list(irrelevant_features))
```

```python
def compute_metrics(threshold):
    irrelevant_features = set()
    for i in range(len(corr_matrix.columns)):
        if abs(corr_matrix.iloc[i, 2]) > threshold:
            colname = corr_matrix.columns[i]
            irrelevant_features.add(colname)

    df_reduced = combined_df.drop(columns=irrelevant_features)
    y = combined_df['review_scores_rating']
    X = df_reduced.drop(columns=['review_scores_rating'])
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1234)

    model = LinearRegression()
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    
    mse = mean_squared_error(y_test, y_pred)
    rmse = np.sqrt(mse)
    r2 = r2_score(y_test, y_pred)
    
    param_grid = {
        'alpha': [0.01, 0.1, 1, 10]
    }

    # Perform Grid Search and find the best hyper parameter
    ridge_model = Ridge()
    grid_search = GridSearchCV(ridge_model, param_grid, cv=5, scoring='neg_mean_squared_error')
    grid_search.fit(X_train, y_train)

    best_ridge_model = grid_search.best_estimator_
    y_pred_ridge = best_ridge_model.predict(X_test)
    
    mse_ridge = mean_squared_error(y_test, y_pred_ridge)
    rmse_ridge = np.sqrt(mse_ridge)
    r2_ridge = r2_score(y_test, y_pred_ridge)
    print(best_ridge_model)
    
    return {
        "threshold": threshold,
        "mse": mse,
        "rmse": rmse,
        "r2": r2,
        "mse_ridge": mse_ridge,
        "rmse_ridge": rmse_ridge,
        "r2_ridge": r2_ridge
    }

# Searching for best correlation cut off threshold
thresholds = np.arange(0.08, 1.0, 0.05)
metrics = [compute_metrics(threshold) for threshold in thresholds]
metrics_df = pd.DataFrame(metrics)
plt.figure(figsize=(14, 7))

# Graph
plt.subplot(2, 2, 1)
plt.plot(metrics_df['threshold'], metrics_df['mse'], label='Linear Regression MSE', marker='o')
plt.plot(metrics_df['threshold'], metrics_df['mse_ridge'], label='Ridge Regression MSE', marker='x')
plt.xlabel('Threshold')
plt.ylabel('MSE')
plt.legend()
plt.grid(True)

plt.subplot(2, 2, 2)
plt.plot(metrics_df['threshold'], metrics_df['rmse'], label='Linear Regression RMSE', marker='o')
plt.plot(metrics_df['threshold'], metrics_df['rmse_ridge'], label='Ridge Regression RMSE', marker='x')
plt.xlabel('Threshold')
plt.ylabel('RMSE')
plt.legend()
plt.grid(True)

plt.subplot(2, 2, 3)
plt.plot(metrics_df['threshold'], metrics_df['r2'], label='Linear Regression R^2', marker='o')
plt.plot(metrics_df['threshold'], metrics_df['r2_ridge'], label='Ridge Regression R^2', marker='x')
plt.xlabel('Threshold')
plt.ylabel('R^2')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```
